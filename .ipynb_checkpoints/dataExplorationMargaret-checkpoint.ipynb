```python
import matplotlib.pyplot as plt # "aliasing" as plt--saves us from typing out the entire library name
import numpy as np
import pandas as pd
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
from pandas import Series, DataFrame
plt.style.use('ggplot')
```

```python
# read in data
spotify = pd.read_csv('data_recent.csv')
spotify
```

```python
# plotting some data
plt.plot(spotify['acousticness'],spotify['popularity'],"bo")
plt.show()
```

```python
# doing subplots
fig, axs = plt.subplots(2,2) # remember matplotlib is imported as plt
#axes = pltAxes.ravel()
axs[0,0].plot(spotify['acousticness'],spotify['popularity'],"bo")
axs[0,1].plot(spotify['danceability'],spotify['popularity'],"bo")
axs[1,0].plot(spotify['loudness'],spotify['popularity'],"bo")
axs[1,1].plot(spotify['instrumentalness'],spotify['popularity'],"bo")
```
